```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {'value': [7, 6, 1, 8, 10, 5, 5, 2, 4, 8]}
df = pd.DataFrame(data, index=['Day 1','Day 2','Day 3','Day 4','Day 5','Day 6','Day 7','Day 8','Day 9','Day 10'])

# Plot
ax = df.plot(kind='line', figsize=(10, 6))

# Set x-axis
ax.set_xticks(range(len(df.index)))
ax.set_xticklabels(df.index, rotation=0)  # Set rotation to desired angle if needed\

# Set y-axis
ax.set_yticks([x for x in range(1,11)])

plt.tight_layout()
plt.show()
```
